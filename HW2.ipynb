```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

import warnings

warnings.filterwarnings("ignore")
```

1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.
Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.
Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.
Исключим стоп-слова с помощью stop_words='english'. 
Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью CountVectorizer.get_feature_names().

```python
combine_df = pd.read_pickle('tweeter.pkl')
combine_df.head()
```

```python
cvec = CountVectorizer(max_df=0.9, max_features=1000, stop_words='english')
```

```python
tweet_lemmatized = np.array(combine_df['tweet_lemmatized'].apply(lambda x: ' '.join(x)))
tweet_stemmed = np.array(combine_df['tweet_stemmed'].apply(lambda x: ' '.join(x)))
```

```python
bov_lemmatized = cvec.fit_transform(tweet_lemmatized)
bovs_stemmed = cvec.fit_transform(tweet_stemmed)
```

```python
df_lemmatized = pd.DataFrame(bov_lemmatized.toarray(), columns=cvec.get_feature_names())
df_stemmed = pd.DataFrame(bovs_stemmed.toarray(), columns=cvec.get_feature_names())
display(df_lemmatized.sample(3), df_stemmed.sample(3))
```

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(trainDF['text'], trainDF['label'])
```

```python
classifier = linear_model.LogisticRegression()

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
```

```python
models = []

models.append(('CountVectorizer',
               CountVectorizer(analyzer='word',
                               token_pattern=r'\w{1,}',
                               stop_words='english')))
models.append(('TfidfVectorizer',
               TfidfVectorizer(analyzer='word',
                               token_pattern=r'\w{1,}',
                               stop_words='english')))
```

```python
from tqdm import tqdm
```

```python
results = []

for name, model in tqdm(models):
  model.fit(trainDF['text'])
  X_train_count = model.transform(X_train)
  X_test_count = model.transform(X_test)
  classifier.fit(X_train_count, y_train)
  predictions = classifier.predict(X_test_count)
  f1_score = accuracy_score(y_test, predictions)
  results.append((name, f1_score))

stock_model = pd.DataFrame(results, columns=['model', 'f1_score'])
stock_model.sort_values('f1_score', ascending=False)
```

```python
import itertools
```

```python
models = []
max_df = [1, 0.9, 0.5, 0.1]
max_features = [3000, 1500, 1000, 350]
```

```python
for df, features in itertools.product(max_df, max_features):
  models.append((f'CVec_max_df_{df}_max_f_{features}',
                 CountVectorizer(analyzer='word', token_pattern=r'\w{1,}',
                 max_df=df, max_features=features, stop_words='english')))
  models.append((f'Tfidf_max_df_{df}_max_f_{features}',
                 TfidfVectorizer(analyzer='word',
                 token_pattern=r'\w{1,}', max_df=df,
                 max_features=features, stop_words='english')))
```

```python
for name, model in tqdm(models):
  model.fit(trainDF['text'])
  X_train_count = model.transform(X_train)
  X_test_count = model.transform(X_test)
  classifier.fit(X_train_count, y_train)
  predictions = classifier.predict(X_test_count)
  f1_score = accuracy_score(y_test, predictions)
  results.append((name, f1_score))

models = pd.DataFrame(results, columns=['model', 'f1_score'])
models.sort_values('f1_score', ascending=False)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
models = []

models.append(('Tfidf',TfidfVectorizer(analyzer='word',token_pattern=r'\w{1,}',stop_words='english')))
```

```python
features = [10, 100, 500, 1000]

for name, model in models:
  for n in tqdm(features):

    model.fit(trainDF['text'])
    svd = TruncatedSVD(n_components=n, random_state=42, algorithm='arpack')

    X_train_count = model.transform(X_train)
    X_test_count = model.transform(X_test)

    X_train_count_svd = svd.fit_transform(X_train_count)
    X_test_count_svd = svd.transform(X_test_count)

    classifier.fit(X_train_count_svd, y_train)

    predictions = classifier.predict(X_test_count_svd)

    f1_score = accuracy_score(y_test, predictions)

    results.append((str(name) + '_' + str(n), f1_score))

pca_model = pd.DataFrame(results, columns=['model', 'f1_score'])
pca_model.sort_values('f1_score', ascending=False)
```
